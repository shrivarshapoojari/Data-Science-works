# Linear Interpolation

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# create new x
x_new =np.linspace(1,5,10)
y_new =np.interp(x_new,x,y)
```

```python
y_new
```

```python
plt.scatter(x_new,y_new)
```

# Cubic Interpolation


```python
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
```

```python
plt.scatter(x,y)
```

```python
x_new =np.linspace(1,5,10)
y_new =np.interp(x_new,x,y)
```

```python
plt.scatter(x_new,y_new)
```

```python
# Polynomial
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
```

```python
p=np.polyfit(x,y,2) # 2 signifies degree of polynomial
```

```python
x_new =np.linspace(1,5,10)
y_new =np.polyval(p,x_new)
```

```python
plt.scatter(x_new,y_new)
```
