# Merge
# minimum one column data should be common
## watch pandas 3  43:59 if  not understood

```python
data1=pd.DataFrame({"key1":[1,2,3,4,5],
                    "key2":[1,2,6,4,5],
                    "key3":[1,2,7,4,5],
                   "key4":[1,8,3,4,5]})
```

```python
data2=pd.DataFrame({"key4":[1,2,9,4,5],
                    "key5":[1,2,6,4,5],
                    "key6":[1,2,7,4,5]})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2)
```

```python
# we will get common  values from the column with same name
# like intersection in venn diagram
```

```python
pd.merge(data1,data2,how='left') # takes everything from first dataset and prints then for second dataset if there is a match in key 4 which is common column it will keep the dataset else keep NaN
```

```python
pd.merge(data1,data2,how='right') # same as left takes second dataset
```

```python
pd.merge(data1,data2,how='outer') # takes all data from left and right
```

```python
pd.merge(data1,data2,how='cross') # takes every row from data1 and combines it with everything on data2 
```
