```python
tips=sns.load_dataset('tips')
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip)
```

```python
tips.head()
```

```python
tips['smoker'].value_counts()
```

```python
# hue can be used to group the data
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip ,data=tips,hue='smoker')
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,hue='size')
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,style='size') # gives different sshapes to differnt data
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,hue='time' ,style="size")
```

```python
# day vs bill   to plot categorical data vs numerical
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(x='smoker', y='tip', data=tips)
```

```python
sns.jointplot(x=tips.total_bill, y=tips.tip)
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(iris)
```
