```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)

```

```python
plt.scatter(x,y)
plt.xlabel("Time")
plt.ylabel("Speed")
plt.title("Speed time")
```

```python
# control colors using parameter c
```

```python
plt.scatter(x,y, c='r') # makes color red
plt.xlabel("Time")
plt.ylabel("Speed")
plt.title("Speed time")
```

```python
# alpha used to control intensity
plt.scatter(x,y, c='r', alpha=0.5) # makes color red
plt.xlabel("Time")
plt.ylabel("Speed")
plt.title("Speed time")
```

```python
# marker used to change the syle of dot
```

```python
# alpha used to control intensity
plt.scatter(x,y, c='r', alpha=0.9 , marker="X") # makes color red
plt.xlabel("Time")
plt.ylabel("Speed")
plt.title("Speed time")
```

```python
# apply frid using plt.grid()
plt.scatter(x,y, c='r') # makes color red
plt.xlabel("Time")
plt.ylabel("Speed")
plt.title("Speed time")
plt.grid()
```

```python
# control the size of figure
plt.figure(figsize=(6,2))
plt.scatter(x,y, c='r') # makes color red
plt.xlabel("Time")
plt.ylabel("Speed")
plt.title("Speed time")
plt.grid()
```
